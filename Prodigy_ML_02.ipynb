```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Mall_Customers.csv")
```

```python
X=data.drop(columns=['CustomerID','Gender','Age'],axis=1).values
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1])
plt.title("BEFORE CLUSTERING")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=2)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title("THE ELBOW METHOD GRAPH")
plt.xlabel("Annual Income (k$)")
plt.ylabel("within cluster sum square")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
Y=kmeans.fit_predict(X)
```

```python
k_center=kmeans.cluster_centers_
print(k_center)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[Y==0,0],X[Y==0,1],c='orange',s=50,label='CLUSTER 1')
plt.scatter(X[Y==1,0],X[Y==1,1],c='green',s=50,label='CLUSTER 2')
plt.scatter(X[Y==2,0],X[Y==2,1],c='purple',s=50,label='CLUSTER 3')
plt.scatter(X[Y==3,0],X[Y==3,1],c='brown',s=50,label='CLUSTER 4')
plt.scatter(X[Y==4,0],X[Y==4,1],c='yellow',s=50,label='CLUSTER 5')
plt.scatter(k_center[:,0],k_center[:,1],s=90,c='black',label='CLUSTER CENTROID')
plt.title("GROUP OF CUSTOMERS(AFTER CLUSTERING)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
